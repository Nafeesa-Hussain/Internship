```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service

def scrape_naukri_jobs():
    driver_path = "/usr/local/bin/chromedriver"
    
    service = Service(driver_path)
    driver = webdriver.Chrome(service=service)
    
    driver.get("https://www.naukri.com/")
    
    skill_field = driver.find_element(By.ID, "qsb-keyword-sugg")
    location_field = driver.find_element(By.ID, "qsb-location-sugg")
    
    skill_field.send_keys("Data Analyst")
    location_field.send_keys("Bangalore")
    
    search_button = driver.find_element(By.CLASS_NAME, "btn")
    search_button.click()
    
    driver.implicitly_wait(10)
    
    jobs = driver.find_elements(By.CLASS_NAME, "jobTuple")
    
    job_data = []
    for job in jobs[:10]:
        job_title = job.find_element(By.CLASS_NAME, "title").text
        job_location = job.find_element(By.CLASS_NAME, "loc").text
        company_name = job.find_element(By.CLASS_NAME, "companyTuple").text
        experience_required = job.find_element(By.CLASS_NAME, "experience").text
        
        job_data.append({
            "Job Title": job_title,
            "Job Location": job_location,
            "Company Name": company_name,
            "Experience Required": experience_required
        })
 
    driver.quit()
    
    return job_data

if __name__ == "__main__":
    scraped_data = scrape_naukri_jobs()
    df = pd.DataFrame(scraped_data)
    print(df)
```

### Question 3

```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

def initialize_driver():
    options = webdriver.ChromeOptions()
    options.add_argument('--headless') 
    driver = webdriver.Chrome(options=options)
    return driver

# scrape the job data
def scrape_jobs_data():
    job_data = {
        'Job Title': [],
        'Job Location': [],
        'Company Name': [],
        'Experience Required': []
    }
    
    driver = initialize_driver()
    wait = WebDriverWait(driver, 10)
    
    try:
        # Step 1: Open the webpage
        driver.get("https://www.naukri.com/")
        
        # Step 2: Enter "Data Scientist" in the search field and click the search button
        search_field = wait.until(EC.presence_of_element_located((By.ID, 'qsb-keyword-sugg')))
        search_field.send_keys("Data Scientist")
        search_button = wait.until(EC.element_to_be_clickable((By.CSS_SELECTOR, '.search-btn')))
        search_button.click()
        
        # Step 3: Apply location filter - Delhi/NCR
        location_filter = wait.until(EC.element_to_be_clickable((By.XPATH, "//label[text()='Delhi / NCR']")))
        location_filter.click()
        
        # Step 4: Apply salary filter - 3-6 lakhs
        salary_filter = wait.until(EC.element_to_be_clickable((By.XPATH, "//label[text()='3-6 Lakhs']")))
        salary_filter.click()
        
        # Step 5: Scrape data for the first 10 job results
        job_elements = wait.until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, '.jobTuple')))
        for job_element in job_elements[:10]:
            job_data['Job Title'].append(job_element.find_element_by_css_selector('.title').text)
            job_data['Job Location'].append(job_element.find_element_by_css_selector('.location').text)
            job_data['Company Name'].append(job_element.find_element_by_css_selector('.companyName').text)
            job_data['Experience Required'].append(job_element.find_element_by_css_selector('.experience').text)
    
    finally:
        driver.quit()
    
    return job_data

# Step 6: Create a DataFrame of the scraped data
if __name__ == "__main__":
    scraped_data = scrape_jobs_data()
    df = pd.DataFrame(scraped_data)
    print(df)
```

```python
def initialize_driver():
    options = webdriver.ChromeOptions()
    options.add_argument('--headless')  # Optional
    driver = webdriver.Chrome(options=options)
    return driver

def scrape_page_data(driver):
    reviews_data = []
    reviews_elements = driver.find_elements(By.CSS_SELECTOR, 'div._2kHMtA')
    
    for review_element in reviews_elements:
        try:
            rating = review_element.find_element(By.CSS_SELECTOR, 'div._3LWZlK').text
            review_summary = review_element.find_element(By.CSS_SELECTOR, 'p._2-N8zT').text
            full_review = review_element.find_element(By.CSS_SELECTOR, 'div.qwjRop').text

            reviews_data.append({
                'Rating': rating,
                'Review Summary': review_summary,
                'Full Review': full_review
            })
        except NoSuchElementException:
            continue
    
    return reviews_data

# Function to scrape 100 reviews data for the iPhone 11 phone
def scrape_iphone11_reviews_data():
    driver = initialize_driver()
    wait = WebDriverWait(driver, 10)
    
    try:
        driver.get("https://www.flipkart.com/apple-iphone-11-black-64-gb/product-reviews/itm4e5041ba101fd?pid=MOBFWQ6BXGJCEYNY&lid=LSTMOBFWQ6BXGJCEYNYZXSHRJ&marketplace=FLIPKART")
        
        # Scroll to load all reviews (adjust the scroll duration based on your page content)
        scroll_duration = 2
        last_height = driver.execute_script("return document.body.scrollHeight")
        while len(driver.find_elements(By.CSS_SELECTOR, 'div._2kHMtA')) < 100:
            driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
            wait.until(EC.staleness_of(driver.find_element(By.CSS_SELECTOR, 'div._2kHMtA:last-child')))
            new_height = driver.execute_script("return document.body.scrollHeight")
            if new_height == last_height:
                break
            last_height = new_height
            driver.implicitly_wait(scroll_duration)
        
        # Scrape reviews data
        reviews_data = scrape_page_data(driver)
        
        # Ensure we only get 100 reviews
        if len(reviews_data) > 100:
            reviews_data = reviews_data[:100]
    
    finally:
        driver.quit()
    
    return reviews_data

if __name__ == "__main__":
    scraped_data = scrape_iphone11_reviews_data()
    df = pd.DataFrame(scraped_data)
    print(df)
```

```python
def initialize_driver():
    options = webdriver.ChromeOptions()
    options.add_argument('--headless') 
    driver = webdriver.Chrome(options=options)
    return driver

def scrape_page_data(driver):
    laptops_data = []
    laptops_elements = driver.find_elements(By.CSS_SELECTOR, 'div.s-result-item')
    
    for laptop_element in laptops_elements:
        try:
            title = laptop_element.find_element(By.CSS_SELECTOR, 'span.a-text-normal').text
            ratings = laptop_element.find_element(By.CSS_SELECTOR, 'span.a-icon-alt').get_attribute('innerHTML')
            price = laptop_element.find_element(By.CSS_SELECTOR, 'span.a-price span.a-offscreen').text

            laptops_data.append({
                'Title': title,
                'Ratings': ratings,
                'Price': price
            })
        except:
            continue
    
    return laptops_data

# scraping first 10 laptops data
def scrape_laptops_data():
    driver = initialize_driver()
    wait = WebDriverWait(driver, 10)
    
    try:
        driver.get("https://www.amazon.in/")
        
        # Step 2: Enter "Laptop" in the search field and click the search icon
        search_field = wait.until(EC.presence_of_element_located((By.ID, 'twotabsearchtextbox')))
        search_field.send_keys("Laptop")
        search_button = wait.until(EC.element_to_be_clickable((By.ID, 'nav-search-submit-button')))
        search_button.click()
        
        # Step 3: Set CPU Type filter to "Intel Core i7"
        filter_button = wait.until(EC.element_to_be_clickable((By.XPATH, "//span[text()='CPU Type']")))
        filter_button.click()
        intel_i7_filter = wait.until(EC.element_to_be_clickable((By.XPATH, "//span[text()='Intel Core i7']")))
        intel_i7_filter.click()
        apply_button = wait.until(EC.element_to_be_clickable((By.XPATH, "//span[text()='Apply']")))
        apply_button.click()
        
        # Step 4 to 6: Scrape data for the first 10 laptops
        total_laptops_data = []
        while len(total_laptops_data) < 10:
            laptops_data = scrape_page_data(driver)
            total_laptops_data.extend(laptops_data)
            
            try:
                next_page_button = driver.find_element(By.CSS_SELECTOR, 'li.a-last a')
                if next_page_button.get_attribute('aria-disabled') == 'true':
                    break
                next_page_button.click()
            except:
                break

    finally:
        driver.quit()
    
    return total_laptops_data

if __name__ == "__main__":
    scraped_data = scrape_laptops_data()
    df = pd.DataFrame(scraped_data)
    print(df)
```

### Question 8

```python
def initialize_driver():
    options = webdriver.ChromeOptions()
    options.add_argument('--headless')  # Optional
    driver = webdriver.Chrome(options=options)
    return driver

def scrape_page_data(driver):
    quotes_data = []
    quotes_elements = driver.find_elements(By.CSS_SELECTOR, 'div.quote')
    
    for quote_element in quotes_elements:
        try:
            quote_text = quote_element.find_element(By.CSS_SELECTOR, 'a.title').text
            author = quote_element.find_element(By.CSS_SELECTOR, 'a.author').text
            quote_type = quote_element.find_element(By.CSS_SELECTOR, 'div.aq-qc a').text

            quotes_data.append({
                'Quote': quote_text,
                'Author': author,
                'TypeOfQuote': quote_type
            })
        except:
            continue
    
    return quotes_data

# Function to scrape Top 1000 Quotes of All Time
def scrape_top_1000_quotes_data():
    driver = initialize_driver()
    wait = WebDriverWait(driver, 10)
    
    try:
        driver.get("https://www.azquotes.com/")
        
        # Step 2: Click on Top Quotes
        top_quotes_button = wait.until(EC.element_to_be_clickable((By.XPATH, "//a[contains(text(), 'Top Quotes')]")))
        top_quotes_button.click()
        
        # Step 3: Scrape data for the Top 1000 Quotes
        total_quotes_data = []
        while len(total_quotes_data) < 1000:
            quotes_data = scrape_page_data(driver)
            total_quotes_data.extend(quotes_data)
            
            try:
                next_page_button = driver.find_element(By.CSS_SELECTOR, 'div.pagination a.next_page')
                if 'disabled' in next_page_button.get_attribute('class'):
                    break
                next_page_button.click()
            except:
                break

    finally:
        driver.quit()
    
    return total_quotes_data

# dataframe creation
if __name__ == "__main__":
    scraped_data = scrape_top_1000_quotes_data()
    df = pd.DataFrame(scraped_data)
    print(df)
```

```python
def scrape_prime_ministers_data(driver):
    prime_ministers_data = []
    rows = driver.find_elements(By.CSS_SELECTOR, 'table.list-table tbody tr')
    
    for row in rows:
        columns = row.find_elements(By.CSS_SELECTOR, 'td')
        name = columns[0].text
        born_dead = columns[1].text
        term_of_office = columns[2].text
        remarks = columns[3].text

        prime_ministers_data.append({
            'Name': name,
            'Born-Dead': born_dead,
            'Term of Office': term_of_office,
            'Remarks': remarks
        })
    
    return prime_ministers_data

# scraping the data of respected former Prime Ministers of India
def scrape_former_pm_data():
    driver = initialize_driver()
    wait = WebDriverWait(driver, 10)
    
    try:
        driver.get("https://www.jagranjosh.com/")
        
        # Step 2: Click on the GK option
        gk_option = wait.until(EC.element_to_be_clickable((By.LINK_TEXT, 'GK')))
        gk_option.click()
        
        # Step 3: Click on the List of all Prime Ministers of India
        pm_list_link = wait.until(EC.element_to_be_clickable((By.LINK_TEXT, 'List of all Prime Ministers of India')))
        pm_list_link.click()
        
        # Step 4: Scrape the data and create the DataFrame
        prime_ministers_data = scrape_prime_ministers_data(driver)
    
    finally:
        driver.quit()
    
    return prime_ministers_data

if __name__ == "__main__":
    scraped_data = scrape_former_pm_data()
    df = pd.DataFrame(scraped_data)
    print(df)
```

```python
def initialize_driver():
    options = webdriver.ChromeOptions()
    options.add_argument('--headless')  
    driver = webdriver.Chrome(options=options)
    return driver

def scrape_cars_data(driver):
    cars_data = []
    cars_elements = driver.find_elements(By.CSS_SELECTOR, 'div.c-listing-item')
    
    for car_element in cars_elements:
        try:
            name = car_element.find_element(By.CSS_SELECTOR, 'h3.c-listing-title').text
            price = car_element.find_element(By.CSS_SELECTOR, 'div.c-price').text

            cars_data.append({
                'Car Name': name,
                'Price': price
            })
        except:
            continue
    
    return cars_data

def scrape_most_expensive_cars_data():
    driver = initialize_driver()
    wait = WebDriverWait(driver, 10)
    
    try:
        driver.get("https://www.motor1.com/")
        
        # Step 2: Type in the search bar '50 most expensive cars' and click search
        search_bar = wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, 'input.c-header-search__input')))
        search_bar.send_keys('50 most expensive cars')
        search_button = wait.until(EC.element_to_be_clickable((By.CSS_SELECTOR, 'button.c-header-search__button')))
        search_button.click()
        
        # Step 3: Click on '50 most expensive cars in the world.'
        link = wait.until(EC.element_to_be_clickable((By.XPATH, "//a[text()='50 most expensive cars in the world.']")))
        link.click()
        
        cars_data = scrape_cars_data(driver)
    
    finally:
        driver.quit()
    
    return cars_data

if __name__ == "__main__":
    scraped_data = scrape_most_expensive_cars_data()
    df = pd.DataFrame(scraped_data)
    print(df)
```
